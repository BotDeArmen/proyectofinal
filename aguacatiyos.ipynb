###COMPARATIVAS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asumiendo que tienes los DataFrames df_conventional y df_organic con la columna 'region_agrupada' y 'Date' (en datetime)

# Función para obtener las N regiones con precio promedio más alto y más bajo (excluyendo TotalUS)
def top_bottom_n_regions(df, n=4):
    df_filtered = df[df['region_agrupada'] != 'TotalUS'].copy()
    df_monthly = df_filtered.groupby(['region_agrupada', df_filtered['Date'].dt.to_period('M')])['AveragePrice'].mean().reset_index()
    avg_price_by_region = df_monthly.groupby('region_agrupada')['AveragePrice'].mean()
    top_n = avg_price_by_region.nlargest(n).index.tolist()
    bottom_n = avg_price_by_region.nsmallest(n).index.tolist()
    return top_n + bottom_n

# Obtener las regiones para convencional y orgánico (excluyendo TotalUS)
top_bottom_conventional = top_bottom_n_regions(df_conventional)
top_bottom_organic = top_bottom_n_regions(df_organic)

# Filtrar los DataFrames para incluir solo las regiones top y bottom y excluir TotalUS
df_conv_filtered = df_conventional[df_conventional['region_agrupada'].isin(top_bottom_conventional) & (df_conventional['region_agrupada'] != 'TotalUS')].copy()
df_org_filtered = df_organic[df_organic['region_agrupada'].isin(top_bottom_organic) & (df_organic['region_agrupada'] != 'TotalUS')].copy()

# Convertir la columna 'Date' a periodo mensual para ambos DataFrames
df_conv_filtered['YearMonth'] = df_conv_filtered['Date'].dt.to_period('M').astype(str)
df_org_filtered['YearMonth'] = df_org_filtered['Date'].dt.to_period('M').astype(str)

# Calcular el precio promedio mensual por región para ambos tipos
avg_price_conv = df_conv_filtered.groupby(['region_agrupada', 'YearMonth'])['AveragePrice'].mean().reset_index()
avg_price_org = df_org_filtered.groupby(['region_agrupada', 'YearMonth'])['AveragePrice'].mean().reset_index()

# Crear la figura y los subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 7), sharey=True)

# Gráfico para Convencionales
sns.lineplot(data=avg_price_conv, x='YearMonth', y='AveragePrice', hue='region_agrupada', ax=axes[0])
axes[0].set_title('Evolución del Precio (Convencional)')
axes[0].set_xlabel('Mes y Año')
axes[0].set_ylabel('Precio Promedio')
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend(title='Región')
axes[0].grid(True)

# Gráfico para Orgánicos
sns.lineplot(data=avg_price_org, x='YearMonth', y='AveragePrice', hue='region_agrupada', ax=axes[1])
axes[1].set_title('Evolución del Precio (Orgánico)')
axes[1].set_xlabel('Mes y Año')
axes[1].set_ylabel('') # Evitar duplicar la etiqueta del eje y
axes[1].tick_params(axis='x', rotation=45)
axes[1].legend(title='Región')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asumiendo que tienes los DataFrames df_conventional y df_organic con las columnas 'Date', 'AveragePrice' y 'region_agrupada'
# y que 'df['Date']' ya está en formato datetime y 'TotalUS' está excluido

def top_n_peak_regions(df, n=4):
    """Encuentra las N regiones con los picos de precio promedio más altos."""
    df_monthly = df.groupby(['region_agrupada', df['Date'].dt.to_period('M')])['AveragePrice'].mean().reset_index()
    peak_prices = df_monthly.groupby('region_agrupada')['AveragePrice'].max()
    top_n_peaks = peak_prices.nlargest(n).index.tolist()
    return top_n_peaks

# Obtener las 4 regiones con los picos más altos para convencional y orgánico
top_peak_conventional = top_n_peak_regions(df_conventional)
top_peak_organic = top_n_peak_regions(df_organic)

# Filtrar los DataFrames para incluir solo las regiones con los picos más altos
df_conv_peaks = df_conventional[df_conventional['region_agrupada'].isin(top_peak_conventional)].copy()
df_org_peaks = df_organic[df_organic['region_agrupada'].isin(top_peak_organic)].copy()

# Convertir la columna 'Date' a periodo mensual para ambos DataFrames
df_conv_peaks['YearMonth'] = df_conv_peaks['Date'].dt.to_period('M').astype(str)
df_org_peaks['YearMonth'] = df_org_peaks['Date'].dt.to_period('M').astype(str)

# Calcular el precio promedio mensual por región para ambos tipos (para graficar)
avg_price_conv_peaks = df_conv_peaks.groupby(['region_agrupada', 'YearMonth'])['AveragePrice'].mean().reset_index()
avg_price_org_peaks = df_org_peaks.groupby(['region_agrupada', 'YearMonth'])['AveragePrice'].mean().reset_index()

# Crear la figura y los subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 7), sharey=True)

# Gráfico para Convencionales (regiones con picos más altos)
sns.lineplot(data=avg_price_conv_peaks, x='YearMonth', y='AveragePrice', hue='region_agrupada', ax=axes[0])
axes[0].set_title('Evolución del Precio (Convencional - Picos más Altos)')
axes[0].set_xlabel('Mes y Año')
axes[0].set_ylabel('Precio Promedio')
axes[0].tick_params(axis='x', rotation=45) # Eliminado ha='right'
axes[0].legend(title='Región')
axes[0].grid(True)

# Gráfico para Orgánicos (regiones con picos más altos)
sns.lineplot(data=avg_price_org_peaks, x='YearMonth', y='AveragePrice', hue='region_agrupada', ax=axes[1])
axes[1].set_title('Evolución del Precio (Orgánico - Picos más Altos)')
axes[1].set_xlabel('Mes y Año')
axes[1].set_ylabel('')
axes[1].tick_params(axis='x', rotation=45) # Eliminado ha='right'
axes[1].legend(title='Región')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

###Continuando con el mismo tema, haz un codigo de barras para cada df donde se vea el totalvolume por cada rregión(las 13) **trimestralmente**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Asumiendo que tienes los DataFrames df_conventional y df_organic
# con las columnas 'Date', 'Total Volume' y 'region_agrupada'

# Función para crear un gráfico de barras trimestral del Total Volume por región (sin TotalUS)
def plot_quarterly_volume_no_totalus(df, title):
    # Filtrar para excluir 'TotalUS'
    df_filtered = df[df['region_agrupada'] != 'TotalUS'].copy()

    # Convertir la columna 'Date' a datetime
    df_filtered['Date'] = pd.to_datetime(df_filtered['Date'])

    # Extraer el trimestre
    df_filtered['Quarter'] = df_filtered['Date'].dt.to_period('Q')

    # Agrupar por región agrupada y trimestre y sumar el Total Volume
    quarterly_volume = df_filtered.groupby(['region_agrupada', 'Quarter'])['Total Volume'].sum().reset_index()

    # Crear el gráfico de barras
    plt.figure(figsize=(16, 8))
    sns.barplot(x='Quarter', y='Total Volume', hue='region_agrupada', data=quarterly_volume)
    plt.title(f'Volumen Total Trimestral por Región (Sin TotalUS) - {title}')
    plt.xlabel('Trimestre')
    plt.ylabel('Volumen Total')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Región', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Aplicar la función al DataFrame de aguacates convencionales
plot_quarterly_volume_no_totalus(df_conventional.copy(), 'Convencional')

# Aplicar la función al DataFrame de aguacates orgánicos
plot_quarterly_volume_no_totalus(df_organic.copy(), 'Orgánico')
```
